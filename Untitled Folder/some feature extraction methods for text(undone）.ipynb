```python
from gensim import corpora, models, similarities
import gensim
stopwords=pd.read_csv("C:/NLP_project/data/stopwords.txt",index_col=False,quoting=3,sep="\t",names=['stopword'], encoding='utf-8')
stopwords=stopwords['stopword'].values

import jieba
import pandas as pd
df = pd.read_csv("C:/NLP_project/data/technology_news.csv", encoding='utf-8')
df = df.dropna()
lines=df.content.values.tolist()

sentences=[]
for line in lines:
    try:
        segs=jieba.lcut(line)
        for item in segs:
            if len(item) > 1 and item not in stopwords:
                sentences.append(item)
        sentences.append(segs)
    except:
        break
```
